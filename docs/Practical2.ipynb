The hypsometric  curve  illustrates the  distribution  of  topography  and  bathymetry  on  Earth.
It  is provided as an external data file, measured in Area (in MKm2). 

![title](./src/prac2Figure1.JPG)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# read in data
with open("./src/hipsometric_curve.dat") as f:
    elevation = f.read().splitlines()

# put in a numpy array and convert to floats(decimal numbers) and kms
elevation = np.array(elevation)
elevation = elevation.astype(np.float)
elevation = elevation*1e-3

# PLOT
plt.plot(elevation)
plt.xlim(0, 520)
plt.xlabel('Area [M km^2]')
plt.ylabel('Elevation [km]')
plt.title('Elevation (in km) vs. Area (in Million km^2)');
```

### Plot the following  
**1)** The curve in the three domains Continental, Oceanic and Trenches in %.  
**2)** A histogram representing the mode (the value mostly represented)

```python
Cont_Elevation = elevation[elevation>-1]
Percent_Continent = 100.0*len(Cont_Elevation)/len(elevation)

Oceanic_Elevation = elevation[elevation<-1]
Oceanic_Elevation = Oceanic_Elevation[Oceanic_Elevation>=-6]
Percent_Oceanic = 100.0*len(Oceanic_Elevation)/len(elevation)

Trenches_Elevation = elevation[elevation<=-6]
Percent_Trenches = 100.0*len(Trenches_Elevation)/len(elevation)

# PLOTS
plt.figure(figsize=(15, 6))

plt.subplot(1, 5, 1)
plt.plot(Cont_Elevation)
plt.xlabel('Area [M km^2]')
plt.ylabel('Elevation [km]');
plt.title('Continental Elevation')

plt.subplot(1, 5, 3)
plt.plot(Oceanic_Elevation)
plt.xlabel('Area [M km^2]')
plt.ylabel('Elevation [km]')
plt.title('Oceanic Elevation')

plt.subplot(1, 5, 5)
plt.plot(Trenches_Elevation)
plt.xlabel('Area [M km^2]')
plt.ylabel('Elevation [km]')
plt.title('Trenches Elevation');
```
